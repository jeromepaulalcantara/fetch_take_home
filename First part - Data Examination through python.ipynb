### Task 2b. Data Quality Checks

```python
#check for missing values per column
print('-----USERS-----')
print(df_users.isnull().sum())
print('-----TRANSACTIONS-----')
print(df_transactions.isnull().sum())
print('-----PRODUCTS-----')
print(df_products.isnull().sum())
```

```python
#Check for any non-NaN string but empty or blank
# Convert BARCODE to string
df_transactions['BARCODE'] = df_transactions['BARCODE'].astype(str)
df_products['BARCODE'] = df_products['BARCODE'].astype(str)

# Check for blank spaces or special characters in the datasets
def check_blank_spaces(df, columns):
    for col in columns:
        # Check for blank spaces or empty strings (strip and check for '')
        blank_check = df[col].str.strip() == '' 
        print(f"Blank space check for {col}:\n", blank_check.sum(), "blank spaces found.\n")


# Apply function to each dataset
user_columns = ['BIRTH_DATE', 'STATE', 'LANGUAGE', 'GENDER']
transaction_columns = ['STORE_NAME', 'USER_ID', 'BARCODE', 'FINAL_QUANTITY', 'FINAL_SALE']
product_columns = ['CATEGORY_1', 'CATEGORY_2', 'CATEGORY_3', 'CATEGORY_4', 'MANUFACTURER', 'BRAND', 'BARCODE']

check_blank_spaces(df_users, user_columns)
check_blank_spaces(df_transactions, transaction_columns)
check_blank_spaces(df_products, product_columns)

```

```python
# Check for non-ASCII characters in the datasets
def check_non_ascii(df, columns):
    # Pattern to detect non-ASCII characters (anything outside the ASCII range)
    pattern = '[^\x00-\x7F]'  # Match characters that are outside the ASCII range (0x00 to 0x7F)
    
    for col in columns:
        if df[col].dtype == 'object':  # Only apply to string columns
            non_ascii_check = df[col].str.contains(pattern, na=False)  # Ignore NaNs
            print(f"Non-ASCII character check for {col}:\n", non_ascii_check.sum(), "non-ASCII characters found.\n")

# Apply function to each dataset
check_non_ascii(df_users, user_columns)
check_non_ascii(df_transactions, transaction_columns)
check_non_ascii(df_products, product_columns)
```

```python
# Check for non-ASCII characters and display the rows that contain them
def show_non_ascii_rows(df, columns):
    # Pattern to detect non-ASCII characters (anything outside the ASCII range)
    pattern = '[^\x00-\x7F]'  # Match characters that are outside the ASCII range (0x00 to 0x7F)
    
    for col in columns:
        if df[col].dtype == 'object':  # Only apply to string columns
            # Identify rows with non-ASCII characters
            non_ascii_rows = df[df[col].str.contains(pattern, na=False)]  # Ignore NaNs
            if not non_ascii_rows.empty:
                print(f"Rows with non-ASCII characters in {col}:")
                print(non_ascii_rows[[col]])  # Show only the problematic column
                print("\n" + "-"*50 + "\n")

# Apply function to each dataset
show_non_ascii_rows(df_users, user_columns)
show_non_ascii_rows(df_transactions, transaction_columns)
show_non_ascii_rows(df_products, product_columns)
```

```python
# Check for duplicates based on unique identifiers, excluding NaNs
print(df_users.dropna(subset=['ID']).duplicated(subset=['ID']).sum())
print(df_transactions.dropna(subset=['RECEIPT_ID']).duplicated(subset=['RECEIPT_ID']).sum())
print(df_products.dropna(subset=['BARCODE']).duplicated(subset=['BARCODE']).sum())
```

```python
# Filter out stores with less than 100 transactions
min_transactions = 100
frequent_stores = df_transactions["STORE_NAME"].value_counts()
frequent_stores = frequent_stores[frequent_stores > min_transactions]

#Barplot to show the distribution of transactions per stores (only for those who have more than 100 transactions)
plt.figure(figsize=(20, 6))
sns.barplot(x=frequent_stores.index, y=frequent_stores.values)
plt.xticks(rotation=45)
plt.title(f"Stores with More Than {min_transactions} Transactions")
plt.xlabel("Store Name")
plt.ylabel("Transaction Count")
plt.show()
```

```python
# Count the number of transactions per store
store_transaction_count = df_transactions['STORE_NAME'].value_counts()

# Filter stores with more than 100 transactions
stores_over_100_transactions = store_transaction_count[store_transaction_count > 100].index

# Filter the transactions dataframe to only include those stores
filtered_transactions = df_transactions[df_transactions['STORE_NAME'].isin(stores_over_100_transactions)]

# Calculate the mean/average final sale by store (filtered for stores with more than 100 transactions)
mean_sale_by_store_filtered = filtered_transactions.groupby('STORE_NAME')['FINAL_SALE'].mean()

# Create a bar plot for average final sale by store
plt.figure(figsize=(20, 6))
sns.barplot(x=mean_sale_by_store_filtered.index, y=mean_sale_by_store_filtered.values, palette='viridis')
plt.xticks(rotation=45)
plt.title("Mean Final Sale by Store (More than 100 transactions)")
plt.xlabel("Store Name")
plt.ylabel("Mean Final Sale")
plt.show()
```

```python
# Calculate the IQR for FINAL_SALE
Q1_sale = df_transactions['FINAL_SALE'].quantile(0.25)
Q3_sale = df_transactions['FINAL_SALE'].quantile(0.75)
IQR_sale = Q3_sale - Q1_sale

# Define the upper and lower bounds for FINAL_SALE
lower_bound_sale = Q1_sale - 1.5 * IQR_sale
upper_bound_sale = Q3_sale + 1.5 * IQR_sale

# Filter out the outliers from FINAL_SALE
df_transactions_cleaned_sale = df_transactions[(df_transactions['FINAL_SALE'] >= lower_bound_sale) & (df_transactions['FINAL_SALE'] <= upper_bound_sale)]

# Plotting a bar plot for FINAL_SALE
plt.figure(figsize=(12, 6))
sns.histplot(df_transactions_cleaned_sale['FINAL_SALE'], kde=True, color='blue', bins=30)
plt.title("Distribution of FINAL_SALE")
plt.xlabel("Final Sale")
plt.ylabel("Frequency")
plt.show()
```

```python
# Calculate the IQR for FINAL_QUANTITY
Q1_quantity = df_transactions['FINAL_QUANTITY'].quantile(0.25)
Q3_quantity = df_transactions['FINAL_QUANTITY'].quantile(0.75)
IQR_quantity = Q3_quantity - Q1_quantity

# Define the upper and lower bounds for FINAL_QUANTITY
lower_bound_quantity = Q1_quantity - 1.5 * IQR_quantity
upper_bound_quantity = Q3_quantity + 1.5 * IQR_quantity

# Filter out the outliers from FINAL_QUANTITY
df_transactions_cleaned_quantity = df_transactions[(df_transactions['FINAL_QUANTITY'] >= lower_bound_quantity) & (df_transactions['FINAL_QUANTITY'] <= upper_bound_quantity)]

# Plot histogram for FINAL_QUANTITY after removing outliers
plt.figure(figsize=(10, 6))
sns.histplot(df_transactions_cleaned_quantity['FINAL_QUANTITY'], kde=True, bins=30)
plt.title('Distribution of FINAL_QUANTITY (After Removing Outliers)')
plt.xlabel('FINAL_QUANTITY')
plt.ylabel('Frequency')
plt.show()
```
